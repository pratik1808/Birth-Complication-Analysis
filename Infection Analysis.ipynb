```python
# Logistic Regression for Infections
from patsy import dmatrices
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import warnings
warnings.filterwarnings('ignore')
# replacing 9 value with NaN. 9 indicates unknown
births['No_Infections_Reported'] = births['No_Infections_Reported'].replace(9, value=np.nan)
births['No_Infections_Reported'].value_counts()


formula = 'No_Infections_Reported ~ 0 + C(Attendant_at_Birth) + C(Birth_Place) + Body_Mass_Index + Cigarettes_Before_Pregnancy' + \
           ' + Birth_Weigth_Grams + C(Marital_Status) + Delivery_Weight + C(Induction_of_Labor) + Mother_Age + Admit_to_Intensive_Care' +\
            '+ Month_Prenatal_Care_Began + Number_of_Prenatal_Visits + Prior_Births_Now_Living + Pre_pregnancy_Weight_lbs' +\
            '+ C(Delivery_Method_Recode) + Previous_Cesarean + Number_of_Previous_Cesareans + Sex_of_Infant'

# setting up the matrix
Y, X = dmatrices(formula, births, return_type='dataframe')
y = Y['No_Infections_Reported'].values

# train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1)

# running the logistic regression
model = LogisticRegression()
result = model.fit(X_train, y_train)

# using model to predict train
prediction_train = model.predict(X_train)
print(metrics.accuracy_score(y_train, prediction_train))

# using model to predict test
prediction_test = model.predict(X_test)
print(metrics.accuracy_score(y_test, prediction_test))

# feature coefficients
weights = Series(model.coef_[0], # model.coef_[0] is necessary because the output is a  list, and this is how to access it
                 index=X.columns.values)
weights.sort_values()

```

# Conclusion
We are not able to beat the accuracy of the baseline where all births are assumed to have no infection. However, we can use our model to identify those patients who may be at a higher-risk of pregnancy using the logistic regression model. This can lead to proactive action during the delivery and immediately thereafter to try and prevent infection. This can be especially practical if medical professionals can identify any of the variables that are the largest contributors to increasing the chance of an infeciton. These variables are Cigarettes before pregnancy, marital status, month prenatal care began, number of previous cesareans, and induction of labor. The most interesting variable in this group is the marital status because it is not necessarily intuitive that this should matter when predicting whether an infection will occur during birth. As noted in the article by St.Germain et al. there is confirmed evidence that if a mother is unmarried there is a higher chance of infection during birth.
